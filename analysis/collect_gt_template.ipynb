```python
import pandas as pd
```

```python
df = pd.read_csv("full_data_dump_8B.csv")
```

```python
from llm_bridge import LLMBridge
```

```python
bridge = LLMBridge(
        model="gpt-5.4-2026-03-05",
        system="You are a concise science communicator.",
        temperature=0.5
    )
```

```python
print(bridge)
```

```python
# Load the prompt template
with open("prompt.txt", "r") as f:
    prompt_template = f.read()

# Create batch of prompts and map each prompt to its query_id
batch_prompts = []
ids = []

for _, row in df.iterrows():
    prompt = prompt_template.replace("{QUERY}", row["query"])
    batch_prompts.append(prompt)
    ids.append(row["id"])
    
# Display the count
print(f"Created {len(batch_prompts)} prompts")
print(f"Created mapping for {len(ids)} prompt->query_id")
```

```python
print(ids[0])
```

```python
answers = bridge.ask_llm_batch(batch_prompts)
```

```python
for id_, a in zip(ids, answers):
    df.loc[df["id"] == id_, "response"] = a
```
