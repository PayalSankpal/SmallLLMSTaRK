```python
import csv
import argparse
import sys
import pandas as pd

def load_csv(path):
    csv.field_size_limit(sys.maxsize)

    df = pd.read_csv(path)

    required = ["query_id", "query_text", "recall@20_vss_merged"]
    for col in required:
        if col not in df.columns:
            raise ValueError(f"{path} missing column: {col}")

    return df.set_index("query_id")
```
